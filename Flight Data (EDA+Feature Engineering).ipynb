## Feature Engineering

```python
final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x:int(x.split('/')[0]))
final_df['Month'] = final_df['Date_of_Journey'].apply(lambda x:int(x.split('/')[1]))
final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x:int(x.split('/')[2]))
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head(2)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
final_df.head(2)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x:int(x.split(':')[0]))
final_df['Arrival_mins'] = final_df['Arrival_Time'].apply(lambda x:int(x.split(':')[1]))
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x:int(x.split(':')[0]))
final_df['Dep_mins'] = final_df['Dep_Time'].apply(lambda x:int(x.split(':')[1]))
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.Total_Stops.unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'NaN':1})
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```

```python
drp_idx = final_df[final_df['Duration']=='5m']
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['duration_hour'] = final_df['Duration'].apply(lambda x:int(x.split(" ")[0].split('h')[0]))
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```
