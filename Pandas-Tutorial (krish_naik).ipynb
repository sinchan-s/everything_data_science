### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Creating a DataFrame

```python
df = pd.DataFrame(np.arange(0,20).reshape(5,4),index=['Row1','Row2','Row3','Row4','Row5'],columns=['Column1','Column2','Column3','Column4'])
```

```python
df
```

### Converting DataFrame(DF) to a CSV file

```python
df.to_csv('Test1.csv')
```

### Selecting two columns as DF

```python
df[['Column1','Column2']]
```

### Using loc

```python
df.loc['Row1']
```

### Using iloc

```python
df.iloc[0:3,0:2]
```

### Converting DF to array

```python
df.iloc[0:3,0:2].values
```
