## RegPlot

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
diamonds2 = diamonds.sample(n=200, random_state=44)
diamonds2
```

```python
sns.set_style('dark')
```

```python
sns.regplot(x='carat', y='price', data=diamonds2)
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, fit_reg=False)
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, scatter=False)
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, ci=None)
```

```python
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}
```

```python
diamonds2['cut_value'] = diamonds2.cut.map(cut_map)
```

```python
diamonds2['cut_value'] = diamonds2['cut_value'].cat.as_ordered()
```

```python
diamonds2.cut_value.astype(int)
```

```python
sns.regplot(x='cut_value', y='price', data=diamonds2)
```

```python
sns.regplot(x='cut_value', y='price', data=diamonds2, x_estimator=np.mean)
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, order=2)
```

### RegPlot: Styling

```python
sns.regplot(x='carat', y='price', data=diamonds2, marker='d')
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, scatter_kws={'s':100, 'alpha':0.5, 'color': 'gray'})
```

```python
sns.regplot(x='carat', y='price', data=diamonds2, ci=None, line_kws={'lw':4, 'linestyle':'-.', 'color': 'green'})
```
